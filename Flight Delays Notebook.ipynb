```python
import pandas as pd
from IPython.display import HTML
```

```python
df = pd.read_csv(r"C:\Users\nopog\OneDrive\Documents\DEPI\Datasets\Flight Delays\flights.csv", dtype = {7: str, 8: str})
```

```python
df.duplicated().sum()
```

```python
df.tail(10)
```

```python
df['DATE'] = pd.to_datetime(df[['YEAR', 'MONTH', 'DAY']])
```

```python
df.head(10)
```

```python
df = df.drop(['WEATHER_DELAY', 'LATE_AIRCRAFT_DELAY', 'AIRLINE_DELAY', 'SECURITY_DELAY', 'AIR_SYSTEM_DELAY', 'CANCELLATION_REASON', 'DAY_OF_WEEK', 'DAY', 'MONTH', 'YEAR'], axis=1)
```

```python
df.head(10)
```

```python
df = df.loc[df['CANCELLED'] != 1]
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace = True)
```

```python
df.isnull().sum()
```

```python
df['SCHEDULED_DEPARTURE'] = df['SCHEDULED_DEPARTURE'].apply(lambda x: f"{int(x):04d}")
df['DEPARTURE_TIME'] = df['DEPARTURE_TIME'].apply(lambda x: f"{int(x):04d}")
df['WHEELS_OFF'] = df['WHEELS_OFF'].apply(lambda x: f"{int(x):04d}")
df['WHEELS_ON'] = df['WHEELS_ON'].apply(lambda x: f"{int(x):04d}")
df['SCHEDULED_ARRIVAL'] = df['SCHEDULED_ARRIVAL'].apply(lambda x: f"{int(x):04d}")
df['ARRIVAL_TIME'] = df['ARRIVAL_TIME'].apply(lambda x: f"{int(x):04d}")
```

```python
df['SCHEDULED_DEPARTURE'] = pd.to_timedelta(df['SCHEDULED_DEPARTURE'].str[:2] + ':' + df['SCHEDULED_DEPARTURE'].str[2:] + ':00')
df['DEPARTURE_TIME'] = pd.to_timedelta(df['DEPARTURE_TIME'].str[:2] + ':' + df['DEPARTURE_TIME'].str[2:] + ':00')
df['WHEELS_OFF'] = pd.to_timedelta(df['WHEELS_OFF'].str[:2] + ':' + df['WHEELS_OFF'].str[2:] + ':00')
df['WHEELS_ON'] = pd.to_timedelta(df['WHEELS_ON'].str[:2] + ':' + df['WHEELS_ON'].str[2:] + ':00')
df['SCHEDULED_ARRIVAL'] = pd.to_timedelta(df['SCHEDULED_ARRIVAL'].str[:2] + ':' + df['SCHEDULED_ARRIVAL'].str[2:] + ':00')
df['ARRIVAL_TIME'] = pd.to_timedelta(df['ARRIVAL_TIME'].str[:2] + ':' + df['ARRIVAL_TIME'].str[2:] + ':00')
```

```python
print(df[['SCHEDULED_DEPARTURE', 'DEPARTURE_TIME', 'WHEELS_OFF', 'WHEELS_ON', 'SCHEDULED_ARRIVAL', 'ARRIVAL_TIME']])
```

```python
df['SCHEDULED_DEPARTURE'] = df['SCHEDULED_DEPARTURE'].astype(str).str[7:]
df['DEPARTURE_TIME'] = df['DEPARTURE_TIME'].astype(str).str[7:]
df['WHEELS_OFF'] = df['WHEELS_OFF'].astype(str).str[7:]
df['WHEELS_ON'] = df['WHEELS_ON'].astype(str).str[7:]
df['SCHEDULED_ARRIVAL'] = df['SCHEDULED_ARRIVAL'].astype(str).str[7:]
df['ARRIVAL_TIME'] = df['ARRIVAL_TIME'].astype(str).str[7:]
```

```python
print(df[['SCHEDULED_DEPARTURE', 'DEPARTURE_TIME', 'WHEELS_OFF', 'WHEELS_ON', 'SCHEDULED_ARRIVAL', 'ARRIVAL_TIME']])
```

```python
df['SCHEDULED_TIME'] = df['SCHEDULED_TIME'] / 60
df['ELAPSED_TIME'] = df['ELAPSED_TIME'] / 60
df['AIR_TIME'] = df['AIR_TIME'] / 60
```

```python
df = df.drop(['DIVERTED', 'CANCELLED'], axis=1)
```

```python
df.head(10)
```

```python
df.to_csv("Flights Cleaned.csv", index=False)
HTML('<a href="Flights Cleaned.csv" download="Flights Cleaned.csv">Click here to download the CSV</a>')
```

### Flights 2

```python
df2 = pd.read_csv(r"C:\Users\nopog\OneDrive\Documents\DEPI\Datasets\Flight Delays\flights.csv", dtype = {7: str, 8: str})
```

```python
df2.head(10)
```

```python
df2['CANCELLATION_REASON'] = df2['CANCELLATION_REASON'].fillna('Not Cancelled')
```

```python
df2['CANCELLED'] = df2['CANCELLED'].replace({0: "Not Cancelled", 1: "Cancelled"})
df2['DIVERTED'] = df2['DIVERTED'].replace({0: "Not DIVERTED", 1: "DIVERTED"})
```

```python
df2.head(10)
```

```python
df2['AIR_SYSTEM_DELAY'] = df2['AIR_SYSTEM_DELAY'].fillna(0)
df2['SECURITY_DELAY'] = df2['SECURITY_DELAY'].fillna(0)
df2['AIRLINE_DELAY'] = df2['AIRLINE_DELAY'].fillna(0)
df2['LATE_AIRCRAFT_DELAY'] = df2['LATE_AIRCRAFT_DELAY'].fillna(0)
df2['WEATHER_DELAY'] = df2['WEATHER_DELAY'].fillna(0)
```

```python
df2.head(10)
```

```python
df2['AIR_SYSTEM_DELAY_YN'] = df2['AIR_SYSTEM_DELAY'].apply(lambda x: 'Yes' if x > 0 else 'No')
df2['SECURITY_DELAY_YN'] = df2['SECURITY_DELAY'].apply(lambda x: 'Yes' if x > 0 else 'No')
df2['AIRLINE_DELAY_YN'] = df2['AIRLINE_DELAY'].apply(lambda x: 'Yes' if x > 0 else 'No')
df2['LATE_AIRCRAFT_DELAY_YN'] = df2['LATE_AIRCRAFT_DELAY'].apply(lambda x: 'Yes' if x > 0 else 'No')
df2['WEATHER_DELAY_YN'] = df2['WEATHER_DELAY'].apply(lambda x: 'Yes' if x > 0 else 'No')
```

```python
df2['CANCELLATION_REASON'] = df2['CANCELLATION_REASON'].replace('A', 'Airline/Carrier')
df2['CANCELLATION_REASON'] = df2['CANCELLATION_REASON'].replace('B', 'Weather')
df2['CANCELLATION_REASON'] = df2['CANCELLATION_REASON'].replace('C', 'National Air System')
df2['CANCELLATION_REASON'] = df2['CANCELLATION_REASON'].replace('D', 'Security')
```

```python
df2.head(10)
```

```python
df2.to_csv("Flights Cleaned 2.csv", index=False)
HTML('<a href="Flights Cleaned 2.csv" download="Flights Cleaned 2.csv">Click here to download the CSV</a>')
```
